```python
import pandas as pd
import numpy as np
```

```python
demographics = pd.read_csv('us_demographics.csv')
```

```python
demographics
```

```python
pop_data = demographics[['NAME','RACE','POPESTIMATE2016']]
```

```python
pop_data
```

```python
statewise = pop_data.groupby(['NAME','RACE'],as_index=False).sum()
```

```python
statewise
```

```python
dummy = [0]*306
```

```python
dummy
```

```python
statewise['new'] = dummy
```

```python
statewise
```

```python
white = statewise[statewise.apply(lambda x : x['RACE'] == 1,axis=1)]
```

```python
white
```

```python
minority = statewise[statewise.apply(lambda x : x['RACE'] > 1,axis=1)]
```

```python
white = white[['NAME','POPESTIMATE2016']]
```

```python
minority = minority[['NAME','POPESTIMATE2016']]
```

```python
white = white.groupby(['NAME'],as_index=False).sum()
```

```python
minority = minority.groupby(['NAME'],as_index=False).sum()
```

```python
white
```

```python
minority
```

```python
final = pd.merge(white,minority,on='NAME')
```

```python
final
```

```python
final = final.rename(columns={'POPESTIMATE2016_x':'white','POPESTIMATE2016_y':'minority'})
```

```python
final
```

```python
final['white_perc'] = (final['white']/(final['minority'] + final['white']))*100
```

```python
final
```

```python
voting = pd.read_excel(r'2016election.xlsx')
```

```python
voting
```

```python
voting['Hillary Clinton'] = voting['Hillary Clinton']*100
```

```python
voting['Donald Trump'] = voting['Donald Trump']*100
```

```python
voting
```

```python
data_df = pd.merge(voting,final,left_on='Votes',right_on='NAME')
```

```python
data_df
```

```python
data_df = data_df[['Votes','Hillary Clinton','Donald Trump','white_perc']]
```

```python
data_df
```

```python
data_df = data_df.rename(columns={'Donald Trump':'donald_trump'})
```

```python
import matplotlib.pyplot as plt
```

```python
x = data_df['white_perc']
```

```python
y = data_df['donald_trump']
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula='donald_trump ~ white_perc',data=data_df).fit()
```

```python
model_prediction = model.predict(x)
```

```python
model_prediction
```

```python
data_df['lr'] = model_prediction
```

```python
data_df
```

```python
model.summary()
```

```python
plt.scatter(x,y)
plt.plot(data_df['white_perc'],data_df['lr'],'r-')
plt.xlabel('White %')
plt.ylabel('Trump vote %')
plt.xlim(0,100)
plt.ylim(0,200)
```
